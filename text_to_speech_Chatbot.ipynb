```python
# Import libraries
import os
from groq import Groq
from gtts import gTTS
import time
from IPython.display import Audio, display

# Set up the Groq client using your API key
client = Groq(api_key=GROQ_API_KEY) #os.environ.get("GROQ_API_KEY"))

# Function to get chatbot response
def chatbot_response(user_input):
    # Create a chat completion request
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": user_input,
            }
        ],
        model="llama3-8b-8192",  # Use your preferred model
    )

    # Get the generated text from the response
    generated_text = chat_completion.choices[0].message.content.strip()  # Strip leading/trailing spaces

    # Convert the response to speech
    audio_file = f"response_{int(time.time())}.mp3"  # Unique audio file name
    tts = gTTS(text=generated_text, lang='en')
    tts.save(audio_file)

    return generated_text, audio_file

# User interaction loop
while True:
    try:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit"]:
            print("Exiting the chat. Goodbye!")
            break

        response, audio_file = chatbot_response(user_input)
        print("Chatbot:", response)

        # Check if the audio file was created and exists
        if audio_file and os.path.exists(audio_file):
            display(Audio(audio_file, autoplay=True))
        else:
            print("Error: Audio file not found or not created.")

    except Exception as e:
        print(f"An error occurred: {e}")
```
